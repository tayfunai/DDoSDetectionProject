## Data Normalization

```python
df.head()
```

```python
df = df.drop(["Unnamed: 0.1"], axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()
```

```python
infinities = np.isinf(df)
infinities.sum().sum()
```

```python
df.isnull().sum().sum()
```

```python
target_value = df["Label"]
input_value = df.drop(["Label"], axis=1)
```

```python
target_value
```

```python
df_scaled = scaler.fit_transform(input_value.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=input_value.columns)
```

```python
norm_df = df_scaled
```

```python
norm_df.isnull().sum().sum()
```

## LinearSVC model training

```python
from sklearn.svm import LinearSVC
linear_svc = LinearSVC(C=1.0, loss='squared_hinge', penalty='l2')
```

```python
from sklearn.model_selection import train_test_split
X = norm_df
y = target_value

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
y
```

```python
import time

svc_start = time.time()
linear_svc.fit(X_train, y_train)
svc_train_time = time.time() - svc_start
```

```python
print(svc_train_time)
```

```python
svc_start = time.time()
y_predict = linear_svc.predict(X_test)
svc_pred_time = time.time() - svc_start
```

```python
print(svc_pred_time)
```

```python
acc_svc = linear_svc.score(X_test, y_test)
```

### LinearSVC model performance

```python
from sklearn.metrics import classification_report 
clr = classification_report(y_test, y_predict)
```

```python
print(clr)
```

## DecisionTree, RandomForest Algorithms

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

```python
dtc_model = DecisionTreeClassifier(criterion='gini', min_samples_split=3)
rfc_model = RandomForestClassifier(criterion='gini', min_samples_split=3)
# abc_model = AdaBoostClassifier(criterion='gini', splitter='best')
```

```python
dtc_start = time.time()
dtc_model.fit(X_train, y_train)
dtc_train_time = time.time() - dtc_start

rfc_start = time.time()
rfc_model.fit(X_train, y_train)
rfc_train_time = time.time() - rfc_start 
```

```python
print(dtc_train_time)
print(rfc_train_time)
```

#### Decision tree and Random forest Classifiers performance

```python
p1 = dtc_model.score(X_test, y_test)
p2 = rfc_model.score(X_test, y_test)

dtc_start = time.time()
y_pred1 = dtc_model.predict(X_test)
dtc_pred_time = time.time() - dtc_start

rfc_start = time.time()
y_pred2 = rfc_model.predict(X_test)
rfc_pred_time = time.time() - rfc_start

print(p1, p2)
print(dtc_pred_time, rfc_pred_time)
```

```python
clr1 = classification_report(y_test,y_pred2)
clr1 = classification_report(y_test,y_pred2)
```
